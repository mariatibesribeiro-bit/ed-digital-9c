```python
def run_quiz():
    print("Bem-vindo ao Quiz da Copa do Mundo!")
    print("Você precisa acertar 10 perguntas consecutivas para ganhar o jogo.")
    print("Se errar uma, o jogo termina.\n")

    # Sortear 10 perguntas do banco
    selected_questions = random.sample(questions, 10)
    correct_count = 0

    for i, q in enumerate(selected_questions):
        print(f"--- Pergunta {i + 1} ---")
        print(q["question"])
        for option_key, option_value in q["options"].items():
            print(f"  ({option_key}) {option_value}")

        user_answer = input("Sua resposta (a ou b): ").lower()

        if user_answer == q["correct_answer"]:
            print("Correto! Próxima pergunta.\n")
            correct_count += 1
        else:
            print(f"Incorreto! A resposta correta era ({q['correct_answer']}) {q['options'][q['correct_answer']]}.\n")
            print("Você perdeu o jogo. Tente novamente!")
            return

    if correct_count == 10:
        print("Parabéns! Você acertou todas as 10 perguntas e ganhou o jogo!")

# Para iniciar o quiz, execute a função:
run_quiz()
```
